# Model predictions

## Preprocessing

```python
tokenizer = AutoTokenizer.from_pretrained("t5-small")

```

```python
prefix = "summarize: "

def preprocess_function(examples):
    inputs = [prefix + doc for doc in examples["article"]]
    model_inputs = tokenizer(inputs, max_length=1024, truncation=True)

    labels = tokenizer(text_target=examples["highlights"], max_length=128, truncation=True)

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_ds = ds.map(preprocess_function, batched=True)
```

# Model

```python
# Set optimizer
from transformers import create_optimizer, AdamWeightDecay

optimizer = AdamWeightDecay(learning_rate=2e-5, weight_decay_rate=0.01)
```

```python
# Load T5 model
from transformers import TFAutoModelForSeq2SeqLM

model = TFAutoModelForSeq2SeqLM.from_pretrained("t5-small")
# Compile model
model.compile(optimizer=optimizer)
```

```python
# Generate batches
from transformers import DataCollatorForSeq2Seq

data_collator = DataCollatorForSeq2Seq(tokenizer=tokenizer, model=model, return_tensors="tf")
generation_data_collator = DataCollatorForSeq2Seq(tokenizer, model=model, return_tensors="tf", pad_to_multiple_of=128)
```

```python
# Change format for training
train_dataset = model.prepare_tf_dataset(
    tokenized_ds["train"],
    batch_size=8,
    shuffle=True,
    collate_fn=data_collator,
)

validation_dataset = model.prepare_tf_dataset(
    tokenized_ds["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=data_collator,
)

generation_dataset = model.prepare_tf_dataset(
    tokenized_ds["validation"],
    batch_size=8,
    shuffle=False,
    collate_fn=generation_data_collator
)
```

```python
# Fit model
model.fit(x=train_dataset, validation_data=validation_dataset, epochs=1)
```

# INFERENCE

```python
test = ds["test"]["article"][:100]
y_test = ds["test"]["highlights"][:100]
```

```python
y_pred = []
for i in tqdm(range(len(test))):
  document = test[i]
  document = "summarize: " + document
  tokenized = tokenizer([document], return_tensors='np')
  out = model.generate(**tokenized, max_length=128)
  with tokenizer.as_target_tokenizer():
    y_pred.append(tokenizer.decode(out[0]))
    print(tokenizer.decode(out[0]))
```

```python
import evaluate
rouge_score = evaluate.load("rouge")
rouge_score.compute(predictions=y_pred, references=y_test)
```
